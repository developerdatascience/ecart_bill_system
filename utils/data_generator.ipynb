```python
# Re-import necessary libraries since execution state was reset
import pandas as pd
import random

# Sample list of Indian FMCG brands and products
brands_products = {
    "Parle": ["Parle-G Biscuits", "Monaco Biscuits", "Hide & Seek Fab", "Krackjack"],
    "Britannia": ["Good Day Butter Cookies", "Marie Gold", "NutriChoice Digestive", "Little Debbie Cake"],
    "Nestle": ["Maggie Noodles", "KitKat", "Everyday Dairy Whitener", "Milkmaid"],
    "Amul": ["Amul Butter", "Amul Cheese Slices", "Amul Taaza Milk", "Amul Ice Cream"],
    "Dabur": ["Dabur Honey", "Dabur Chyawanprash", "Real Fruit Juice", "Dabur Red Toothpaste"],
    "Haldiram's": ["Aloo Bhujia", "Soan Papdi", "Kaju Katli", "Moong Dal"],
    "ITC": ["Bingo! Mad Angles", "Aashirvaad Atta", "Sunfeast Dark Fantasy", "Classmate Notebook"],
    "Patanjali": ["Patanjali Ghee", "Patanjali Dant Kanti", "Patanjali Aloe Vera Juice", "Patanjali Honey"],
    "Colgate": ["Colgate MaxFresh", "Colgate Vedshakti", "Colgate Zig Zag Brush"],
    "Hindustan Unilever": ["Lux Soap", "Dove Shampoo", "Pears Face Wash", "Surf Excel Detergent"],
}

# Define product categories
category_mapping = {
    "Biscuits & Cookies": ["Parle-G Biscuits", "Monaco Biscuits", "Hide & Seek Fab", "Krackjack",
                           "Good Day Butter Cookies", "Marie Gold", "NutriChoice Digestive"],
    "Dairy Products": ["Amul Butter", "Amul Cheese Slices", "Amul Taaza Milk", "Amul Ice Cream"],
    "Instant & Snacks": ["Maggie Noodles", "Bingo! Mad Angles"],
    "Chocolates & Sweets": ["KitKat", "Sunfeast Dark Fantasy", "Soan Papdi", "Kaju Katli"],
    "Beverages": ["Real Fruit Juice", "Patanjali Aloe Vera Juice"],
    "Health & Wellness": ["Dabur Honey", "Dabur Chyawanprash", "Patanjali Honey"],
    "Personal Care": ["Dabur Red Toothpaste", "Colgate MaxFresh", "Colgate Vedshakti", "Colgate Zig Zag Brush",
                      "Lux Soap", "Dove Shampoo", "Pears Face Wash"],
    "Home Essentials": ["Surf Excel Detergent", "Aashirvaad Atta", "Classmate Notebook"],
    "Savories & Namkeen": ["Aloo Bhujia", "Moong Dal"],
}

# Generate a sample dataset
data = []
for brand, products in brands_products.items():
    for product in products:
        mrp = round(random.uniform(10, 500), 2)  # Random price between 10 and 500
        pack_size = random.choice(["100g", "200g", "500g", "1kg", "50ml", "100ml", "250ml", "500ml", "1L", "2L"])
        category = next((cat for cat, items in category_mapping.items() if product in items), "Others")

        # Ensure Biscuits & Cookies have pack_size in grams
        if category == "Biscuits & Cookies":
            pack_size = random.choice(["100g", "200g", "500g", "1kg"])

        data.append([product, brand, mrp, pack_size, category])

# Convert to DataFrame
df = pd.DataFrame(data, columns=["product_name", "brand", "mrp", "pack_size", "category"])
df['product_name-qty'] = df.product_name + '-' + df.pack_size
new_df = df.copy(deep=True)
new_df.drop(columns=['product_name'], axis=1, inplace=True)
col_to_rename = {"product_name-qty":"product_name"}
new_df.rename(columns=col_to_rename, inplace=True)
new_df = new_df[['product_name', 'brand', 'mrp', 'pack_size', 'category']]

# Save to CSV
# csv_filename = "../data/indian_fmcg_products.csv"
# new_df.to_csv(csv_filename, index=False)

# csv_filename
```
